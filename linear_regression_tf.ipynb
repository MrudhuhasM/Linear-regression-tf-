Implenting simple linear regression on radom generated data

```python
data = pd.read_csv('random_data.csv')
```

```python
data.head()
```

```python
s = np.random.uniform(1,100)
i = np.random.uniform(1,100)
```

```python
slope = tf.Variable(s,tf.float32)
intercept = tf.Variable(i,tf.float32)
```

```python
slope,intercept
```

```python
target = data['target'].values
feature = data['feature'].values
```

```python
def linear_regression(intercept,slope,feature):
    return slope*feature + intercept

def loss_function(intercept,slope,feature,target):
    predictions = linear_regression(intercept,slope,feature)
    return tf.keras.losses.mse(target,predictions)
```

```python
opt = tf.keras.optimizers.Adam(1)

for i in range(160):
    opt.minimize(lambda : loss_function(intercept,slope,feature,target) , var_list = [intercept,slope])
    
    if i % 10 == 0:
        print(loss_function(intercept,slope,feature,target).numpy())
```

Implementing Lineas regression on housing prices

```python
data = pd.read_csv('housing_prices.csv')
```

```python
data = dabl.clean(data)
```

```python
data = data.drop(['id','date'],axis=1)
```

```python
data.info()
```

```python
data.head()
```

```python
X = data.drop('price',axis=1).values
y = data['price'].values
X = np.array(X,dtype=float)
```

```python
params = []
for i in range(X.shape[1]):
    params.append(np.random.random())
params.append(np.random.random())

params = np.array(params)
params = params.reshape(1,-1)
params = tf.Variable(params,tf.float32)
```

```python
def linear_regression(params,features=X):
    #return tf.Variable(np.add(np.matmul(features,params[:-1]), params[-1]))
    return tf.math.reduce_sum(np.multiply(X,params[:,:-1]),1) + params[:,-1]
def loss_function(params,features=X,target=y):
    predictions = linear_regression(params)
    return tf.keras.losses.mse(target,predictions)
```

```python
opt = tf.keras.optimizers.Adam(50)

for i in range(1000):
    opt.minimize(lambda : loss_function(params) , var_list = [params])
    if i % 100 == 0:
        print(f'Loss : {loss_function(params,X,y).numpy()} \nParams {params.numpy()[0]}')
        

# opt = tf.keras.optimizers.Adam(1)

# for i in range(160):
#     opt.minimize(lambda : loss_function(intercept,slope,feature,target) , var_list = [intercept,slope])
    
#     if i % 10 == 0:
#         print(loss_function(intercept,slope,feature,target).numpy())
```
